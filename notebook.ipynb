```python
import configparser
import os
import pickle
import dill
import numpy as np
import torch
from GNNExplainer import GraphNetworkExplainer
from graphConvolutionalNetworkClassifier import GraphNetwork
import random
from yfiles_jupyter_graphs import GraphWidget

from utils import custom_graph_element_color_mapping

seed = 0
torch.manual_seed(seed)
random.seed(seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)
torch.backends.cudnn.deterministic = True

if __name__ == '__main__':
    filepaths = os.listdir("./dataset/graph/test_to_show")
    print(filepaths)
    for filename in filepaths:
        with open("./graph_test_to_show/" + filename, 'rb') as file:
            try:
                graph = dill.load(file)
                widget = GraphWidget()
                widget.set_node_color_mapping(
                    lambda node: custom_graph_element_color_mapping(node, graph.min_node_importance, graph.max_node_importance))
                widget.nodes = graph.nodes
                widget.edges = graph.edges

                widget.directed = True
                widget.get_heat_mapping()
                print(filename.split('.')[0] + ".json")
                display(widget)
                print("\n\n\n")
            except:
                print("")
            
            
```
